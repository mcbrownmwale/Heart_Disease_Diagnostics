# Storytelling Data Visualization on Exchange Rate
## Importing the required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use('ggplot')
```

## Read in the Dataset and Inspect the Dataset

```python
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rates.info()
```

## Data Cleaning
### 1. Renaming Columns

```python
new_column_names = []
for column_name in exchange_rates.columns:
    column_name = column_name.replace('[', '')
    column_name = column_name.replace(' ]', '')
    column_name = column_name.replace(' ', '_')
    column_name = column_name.lower()
    new_column_names.append(column_name)
exchange_rates.columns = new_column_names
exchange_rates.rename(columns={'period\\unit:': 'time'}, inplace = True)
exchange_rates.columns
```

### 2. Dropping Unnecessary Columns and Change the Datatype of 'time' column

```python
exchange_rates['time'] = pd.to_datetime(exchange_rates['time'])
euro_to_dollar = exchange_rates[['time', 'us_dollar']]
euro_to_dollar.info()
```

### 3. Changing Index

```python
euro_to_dollar.set_index(keys = 'time', drop = True, inplace = True)
euro_to_dollar.head()
```

### 4. Sorting Records

```python
euro_to_dollar = euro_to_dollar.sort_values(['time'])
euro_to_dollar.head()
```

### 5. Clean `us_dollar` column and change its datatype

```python
euro_to_dollar['us_dollar'].value_counts()
```

```python
euro_to_dollar['us_dollar'] = euro_to_dollar[euro_to_dollar['us_dollar'] != '-'].astype('float').dropna()
euro_to_dollar.info()
```
